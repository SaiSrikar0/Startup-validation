```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Startups.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop('Website',axis=1,inplace=True)
df.drop('Logo',axis=1,inplace=True)
df.drop('Seed-DB / Mattermark Profile',axis=1,inplace=True)
df.drop('Crunchbase / Angel List Profile',axis=1,inplace=True)
df.drop('Office Address',axis=1,inplace=True)
df.drop('Y Combinator Year',axis=1,inplace=True)
df.drop('Y Combinator Session',axis=1,inplace=True)
df.drop('Mapping Location',axis=1,inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Year Founded'].fillna(df['Year Founded'].mode()[0],inplace=True)
df['Description'].fillna(df['Description'].mode()[0],inplace=True)
df['Founders'].fillna(df['Founders'].mode()[0],inplace=True)
df['Categories'].fillna(df['Categories'].mode()[0],inplace=True)
df['Amounts raised in different funding rounds'].fillna(df['Amounts raised in different funding rounds'].mode()[0],inplace=True)
df['Headquarters (City)'].fillna(df['Headquarters (City)'].mode()[0],inplace=True)
df['Headquarters (US State)'].fillna(df['Headquarters (US State)'].mode()[0],inplace=True)
df['Headquarters (Country)'].fillna(df['Headquarters (Country)'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv('Startups_cleaned.csv',index=False)
```
